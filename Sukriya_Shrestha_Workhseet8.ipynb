```python
import numpy as np
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.metrics import  confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/Diabetes.csv')
df
```

```python
X = df.drop(columns = ['Outcome'], axis = 1)
Y = df['Outcome']
print(X.columns)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# split into training and test set(70% training, 30% testing)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled,Y, test_size = 0.3, random_state = 42)
```

**Implement Classification Model**

```python
#initialize decisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)

```

```python
prediction = model.predict(X_test)  #y_pred ho yo predictions bhnya
prediction
```

```python
#initialize RandomForestClassifier
model1 = RandomForestClassifier(n_estimators = 100, random_state=42)
model1.fit(X_train, Y_train)
```

```python
prediction1 = model1.predict(X_test)
prediction1
```

```python
#checking accuracy for decision tree classifier
accuracy = accuracy_score(Y_test, prediction)
accuracy
```

```python
#conf matrix for decision tree classifier
conf_matrix = confusion_matrix(Y_test, prediction)
conf_matrix
```

```python
#checking accuracy for RandomForestClassifier
accuracy1 = accuracy_score(Y_test, prediction1)
accuracy1
```

```python
#conf_matrix for randomforestclassifier
conf_matrix1 = confusion_matrix(Y_test, prediction1)
conf_matrix1
```

```python
#display F1 Scores
dt_f1 = f1_score(Y_test, prediction, average = 'weighted')
rf_f1 = f1_score(Y_test, prediction1, average = 'weighted')
print(f"Decision tree F1 Score: {dt_f1:.2f}")
print(f"Random forest F1 Score: {rf_f1:.2f}")
```

**Cross Validation**(use grid or random CV)
Hyperparameter Tuning:

```python
# parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [100,200,300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
}
```

```python
#Print best parameter and best Cross-validation score
grid_search = GridSearchCV(model1, param_grid, cv = 5, scoring = 'neg_mean_squared_error')
grid_search.fit(X_train, Y_train)

#print the best parameters and best cross-validation score
print("Best parameters: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)
```

**Implement Regression Model**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

**Decision Tree Regressor**

```python
#initialize model forest Regressor
model = DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
#checking accuracy
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Decision Tree Regression Mean Squared Error: ", mse)
print("Decision Tree Regression R squared: ", r2)
```

**Random Forest Regression**

```python
model1 = RandomForestRegressor()
model1.fit(X_train, Y_train)
```

```python
y_pred1 = model1.predict(X_test)
y_pred1
```

```python
#Mean squared error and r-squared
mse = mean_squared_error(Y_test, y_pred1)
r2 = r2_score(Y_test, y_pred1)

print("Random Forest Regression Mean Squared Error: ", mse)
print("Random Forest Regression R squared: ", r2)
```

**HyperParameter Tuning using RandomSearchCV**

```python
rf_grid = {
    'n_estimators': [100, 200, 100],
    'max_depth': [10,20, None],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4]
}
```

```python
random_search = RandomizedSearchCV(model1, param_distributions = rf_grid, cv = 5, scoring ='neg_mean_squared_error')
random_search.fit(X_train, Y_train)
#print the best parameters and best cross-validation score
print("Best parameters: ", random_search.best_params_)
print("Best cross-validation score: ", random_search.best_score_)


```
